```python
import pandas as pd
from bs4 import BeautifulSoup
import requests 
```

```python
#Facebook stock prices

page = requests.get('https://finance.yahoo.com/quote/FB/history?period1=1356994800&period2=1556488800&interval=1d&filter=history&frequency=1d&guccounter=1')
print(page.status_code)

soup=BeautifulSoup(page.text, 'html.parser')
stocks = soup.find_all('tr', class_='BdT Bdc($c-fuji-grey-c) Ta(end) Fz(s) Whs(nw)')
print(stocks[0].text)


#FB stock prices
dates = []
close_price = []
volume = []

for stock in stocks:
    dates.append(stock.find_all('td')[0].text)
    close_price.append(stock.find_all('td')[4].text)
    volume.append(stock.find_all('td')[6].text)

cols = {'Date': dates, 'Close price':close_price, 'Volume':volume}
df_FB=pd.DataFrame(cols)
df_FB.set_index('Date', inplace=True)
df_FB
```
